```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
List=list()

for i in range(1,14):
  url = "https://www.remax.com.tr/konut/kiralik/istanbul-avrupa"+"?page"+str(i)
  
  req = requests.get(url)
  soup = BeautifulSoup(req.text, "html.parser")
  for link in soup.find_all("span",attrs={"class":"portfoy-no"}):
      link=link.text.split(":")[1]
      List.append([link])
print(List)


```

```python
df=pd.DataFrame(columns=["No","Fiyat","Konum","Emlak Tipi","m2 (Brüt)","m2 (Net)","Oda Sayısı","Bina Yaşı","Bulunduğu Kat","Kat Sayısı","Isıtma","Banyo Sayısı","Balkon Sayısı","Eşya Durumu","Kullanım Durumu","Site İçerisinde Mi"])
for i in range(0,len(List)):
  url = "https://www.remax.com.tr/portfoy/"+str(List[i]).strip("[").strip("]").strip("'").strip(" ")
  print(url)
  req = requests.get(url)
  soup = BeautifulSoup(req.text, "html.parser")
  for link in soup.find_all("div",attrs={"class":"properties"}):
      price=link.find("strong").text
      breadcrumbs=link.find("div",attrs={"class":"breadcrumbs"}).text
      listeData=list()
      for li in  link.find_all("span"):
        listeData.append(li.text)
      df= df.append(pd.DataFrame([[List[i], price, breadcrumbs,listeData[1],listeData[2],listeData[3],listeData[4],listeData[5],listeData[6],listeData[7],listeData[8],listeData[9],listeData[10],listeData[11],listeData[12],listeData[13]]],
                                 columns=["No","Fiyat","Konum","Emlak Tipi","m2 (Brüt)","m2 (Net)","Oda Sayısı","Bina Yaşı","Bulunduğu Kat","Kat Sayısı","Isıtma","Banyo Sayısı","Balkon Sayısı","Eşya Durumu","Kullanım Durumu","Site İçerisinde Mi"]))
      print(df)

print(df)
```

```python
df.to_csv("remax.csv", encoding='utf-8', index=False)
```
